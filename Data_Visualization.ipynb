# **VISUALISATION**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import squarify
```

# LINE CHART:
This line chart shows Average Price by Brand

```python
brand_price = df.groupby("Brand")["Price"].mean().sort_values()    #same brand are considered together.
plt.figure(figsize=(18,6))    #set width and height
plt.plot(brand_price.index, brand_price.values, marker="o", color="blue")  #x & y axis data
plt.xticks(rotation=45)  #Tilt the names so they don’t overlap and can be readable 
plt.title("Average Price by Brand")
plt.xlabel("Brand names")
plt.ylabel("Average Price (₹)")
plt.show()
```

# Bar Chart:
Top 10 brands

```python
brand_count = df["Brand"].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=brand_count.index, y=brand_count.values, palette="viridis")
plt.title("Top 10 Brands by Number of Models")
plt.ylabel("Number of Models")
plt.xticks(rotation=45)
plt.show()
```

# Scatter Plot:

Price vs Rating

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="Price", y="Avg Rating", hue="Brand", alpha=0.7)
plt.title("Price vs Average Rating")
plt.show()
```

# Pairplot:
 
 Price, Rating & discount

```python
sns.pairplot(df[["Price","Avg Rating","Discount"]].dropna())
plt.show()
```

# Box Plot:

Price Distribution of top 5 Brands

```python
top_brands = df["Brand"].value_counts().head(5).index
plt.figure(figsize=(14,7))
sns.boxplot(data=df[df["Brand"].isin(top_brands)], x="Brand", y="Price")
plt.title("Price Distribution of Top 5 Brands")
plt.show()
```

# Heatmap:
 higher battery capacity or price leads to higher ratings.

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[["Price","Avg Rating","Discount"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

# Pie Chart:
Top 5 brands shared by market

```python
top5 = df["Brand"].value_counts().head(5)
plt.figure(figsize=(7,7))
plt.pie(top5, labels=top5.index, autopct="%1.1f%%", startangle=140)
plt.title("Top 5 Brands Share")
plt.show()
```

# Violin Plot:
Distribution of ratings per brand in violin shape.

```python
plt.figure(figsize=(12,6))
sns.violinplot(data=df[df["Brand"].isin(top_brands)], x="Brand", y="Avg Rating")
plt.title("Rating Distribution by Brand")
plt.show()
```

# KDE Plot:
Price distribution

```python
plt.figure(figsize=(10,6))
sns.kdeplot(df["Price"].dropna(), fill=True, color="green")
plt.title("Price Distribution")
plt.xlabel("Price (₹)")
plt.show()
```

# Histogram:
Price Distribution

```python
plt.figure(figsize=(10,6))
sns.histplot(df["Price"].dropna(), bins=30, kde=True)
plt.title("Price Distribution")
plt.xlabel("Price (₹)")
plt.show()
```

# Donut Chart: 
RAM Distribution


```python
ram = df["Memory Details"].str.extract("(\d+) GB RAM")[0].value_counts()
plt.figure(figsize=(8,8))
plt.pie(ram, labels=ram.index+" GB RAM", autopct="%1.1f%%", startangle=90)
centre = plt.Circle((0,0),0.70,fc="white")   # donut effect
fig = plt.gcf()
fig.gca().add_artist(centre)
plt.title("RAM Distribution")
plt.show()
```

# TreeMap(Squarify):
Brand Distribution

```python
plt.figure(figsize=(12,6))
squarify.plot(sizes=df["Brand"].value_counts().values, 
              label=df["Brand"].value_counts().index, alpha=.8)
plt.title("Brand Market Share (Treemap)")
plt.axis("off")
plt.show()
```
